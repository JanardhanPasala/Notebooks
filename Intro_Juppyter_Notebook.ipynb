```python
from pandas_datareader import data
from datetime import datetime
# import matplotlib
```

```python
start = datetime(2019, 12, 1)
end = datetime(2019, 12, 31)
```

```python
df = data.get_data_yahoo('tcs', start, end)
df.head(), df.tail()
```

```python
df.info()
```

```python
df
```

```python
df.Close.plot()
```

```python
df.plot(y='Close', color='red')
```

```python
%matplotlib inline
```

```python
df.Close.plot(figsize=(15,15))
```

```python
df[['Close', 'Open']].plot(figsize=(10,7))
```

```python
df.loc['2019-12-13':'2019-12-29', ['Open', 'Close']].plot(figsize=(10,7))
```

```python
df.Open.pct_change().plot()
```
